```python
import numpy as np
```

```python
def Grad(f, xk,h):
    n = xk.size
    g = np.zeros(n)
    for i in range(0, n):
        a = np.copy(xk)
        b = np.copy(xk)
        a[i] -= h
        b[i] += h
        g[i] = (f(a) - f(b))/(2*h)
    return g
```

```python
def Hess(f,xk,h):
    n=xk.size
    H=np.zeros((n,n))
    for i in range(0,n):
        for j in range(0,n):
            ff=np.copy(xk)
            ff[i]+=h
            ff[j]+=h
            
            fb=np.copy(xk)
            fb[i]+=h
            fb[j]-=h
            
            bf=np.copy(xk)
            bf[i]-=h
            bf[j]+=h
            
            bb=np.copy(xk)
            bb[i]-=h
            bb[j]-=h
            
            H[i,j]=(f(ff)-f(fb)-f(bf)+f(bb))/(4*h**2)
    return H
            
```

```python
def es_pos_def(A): #Solo para matrices Simétricas
        try:
            np.linalg.cholesky(A)
            return True
        except np.linalg.LinAlgError:
            return False
        return False
```

```python
def PruebaMod_Hess(Hk, Beta ) :
    n=int(np.sqrt(np.size(Hk)))
    tk=0
    v=np.matrix.diagonal(Hk)
    
    if min(v)<0:
        tk=-min(v)+Beta    
    

    while es_pos_def(Hk+np.eye(n)*tk)==False:
           tk= max(2*tk, Beta);
    
    return Hk+np.eye(n)*tk
     
```

```python
def BactrackSearch(f,xk,pk,gk,alpha0,c,rho):
    alpha_k=alpha0
   

    while f(xk+alpha_k*pk)> f(xk)+c*alpha_k*np.dot(gk,pk):
        
        alpha_k=alpha_k*rho
    return alpha_k
    
```

```python
def rosenbrock(v):
    #a=1 b=100
    return ((1-v[0])**2+100*(v[1]-v[0]**2)**2)
```

```python
def check_optimality(gk,Hk,tolT):
    return es_pos_def(Hk) and all(abs(gk)<tolT)
```
